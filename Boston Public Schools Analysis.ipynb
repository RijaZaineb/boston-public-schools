```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas
import bqplot
import numpy as np
import contextily as ctx
import ipywidgets
```

```python
# reading Boston Council Districts dataset
bostonCouncil = geopandas.read_file("City_Council_Districts.geojson")
```

```python
bostonCouncil
```

```python
bostonCouncil.plot()
```

```python
# Coordinate Reference System of Boston Council District dataset
bostonCouncil.crs
```

```python
fig,ax = plt.subplots(figsize=(4,4))
bostonCouncil.plot(ax=ax, alpha=0.5, edgecolor='k', color='yellow')
ctx.add_basemap(ax=ax, crs=bostonCouncil.crs)
plt.show()
```

```python
# reading Boston Public Schools dataset
schools = geopandas.read_file("Public_Schools.geojson")
```

```python
schools
```

```python
schools.plot()
```

```python
# Coordinate Reference System of Boston Public Schools dataset
schools.crs
```

```python
# looking at different school types in the dataset
schools['SCH_TYPE'].unique()
```

```python
# plotting schools over the council districts
fig, ax = plt.subplots(figsize=(10,10))
bostonCouncil.plot(ax=ax, alpha=0.4, edgecolor='k', color="yellow")
schools.plot(column='SCH_TYPE', ax=ax, cmap='winter', categorical=True, legend=True)

ctx.add_basemap(ax=ax,crs=bostonCouncil.crs)
plt.show()
```

```python
# 1. hardcoding school type

school_type = 'ES'

schools_subset = schools[schools['SCH_TYPE']==school_type]

fig, ax = plt.subplots(figsize=(5,5))

bostonCouncil.plot(ax=ax, alpha=0.4, edgecolor='k', color="yellow")
schools_subset.plot(column='SCH_TYPE', ax=ax, cmap='winter', categorical=True, legend=True)

ctx.add_basemap(ax=ax,crs=bostonCouncil.crs)
plt.show()
```

```python
# 2. Making a function to change school type

def make_geo_plot(school_type):
    schools_subset = schools[schools['SCH_TYPE']==school_type]

    fig, ax = plt.subplots(figsize=(10,10))
   
    bostonCouncil.plot(ax=ax, alpha=0.4, edgecolor='k', color="yellow")
    schools_subset.plot(column='SCH_TYPE', ax=ax, cmap='winter', categorical=True, legend=True)

    ctx.add_basemap(ax=ax,crs=bostonCouncil.crs)
    plt.show()
```

```python
# 2.5 Testing function
make_geo_plot('K-8')
```

```python
# 3 Final Viz - Using widget with the function to allow users to select school type

import ipywidgets
@ipywidgets.interact(Type=schools['SCH_TYPE'].unique())
def make_geo_plot(Type):
    schools_subset = schools[schools['SCH_TYPE']==Type]

    fig, ax = plt.subplots(figsize=(10,10))
  
    bostonCouncil.plot(ax=ax, alpha=0.4, edgecolor='k', color="yellow")
    schools_subset.plot(column='SCH_TYPE', ax=ax, cmap='winter', categorical=True, legend=True)

    ctx.add_basemap(ax=ax,crs=bostonCouncil.crs)
    plt.show()
```

<b>Explanation</b><br>

I have created this visualization to show public school locations in Boston spread across different council districts. I used two datasets to build this vizualization i.e., Boston Council Districts and Boston Public Schools. Both these datasets were taken from Analyze Boston - City of Boston Open Data Hub. I downloaded the GeoJson files that are submitted with this notebook.

Latitude/Longitude are used to display data on map. The coordinate reference system for both the datasets was found using the .crs attribute. Both the datasets had Geographic 2D coordinate reference system with EPSG code 4326. The EPSG number (“European Petroleum Survey Group”) is a code that tells about the coordinate system of the dataset. EPSG code 4326 belongs to the WGS84 (World Geodetic System 1984) coordinate system. It comprises of reference elliposoid, horizontal datum, vertical datum and a coordinate system. Coordinates are in decimal degrees (latitude, longitude). In case of council district, the geometry is Multipolygon while for schools it is Point. Coordinate values of Multipolygon and Point are latitude, longitude values. I decided to use the same projection for the visualization i.e., I did not change the coordinate reference system as this projection is appropriate for this type of data and displays data the way I wanted.

There are different school types like ES (Elementary School), K-8, K-12 etc. in this dataset. I have created a dropdown widget to give the users an option to select school type and the data on the viz gets updated accordingly. The potential users of this visuzalization could be government officials in the department of education or other officials who are interested in learning about how many and what type of public schools are located in different council districts. This could inform their decisions of opening more schools or different type of schools in a certain location. Similarly, this visualization could also be used by residents of Boston city. Parents might be interested in learning where different types of schools or schools of interest are located.

From the visualization we can see that Elementary Schools (ES) are more in number and spread out across council districts while high schools are fewer in comparison. Similarly, K-8 are more thank K-12 schools. It is understandable as it is a general pattern in the US districts that there are fewer high schools than middle schools and fewer middle schools than elementary schools.
